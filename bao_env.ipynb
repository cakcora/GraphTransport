```python
def convert_date(date_string):
    original_date = datetime.strptime(date_string, "%Y-%m-%dT%H:%M:%S.%fZ")
    return original_date.strftime("%Y-%m-%d")

def clean_price_data(network):
    csv_name = "dataset/price_dataset/"+network+"_price.csv"
    txt_file = "dataset/price_dataset/"+network+".txt"

    #Edit csv file to make data usable and store in text file
    with open(csv_name, "r") as file:
        # Read the entire contents of the file into a string
        file_contents = file.read()
        file_contents = file_contents.replace(";" , " ")
        with open(txt_file, "w") as file:
        # Write the modified string to the file
            file.write(file_contents)

    price_data = pd.read_csv(txt_file, sep=' ')

    converted_dates = []
    for date in price_data['timestamp']:
        print(date)
        converted_dates.append(convert_date(date))

    price_data['date'] = converted_dates
    price_data = price_data.drop('timestamp',axis=1)
    price_data.to_csv(csv_name)
    os.remove(txt_file)
    print("Generated cleaned price file:" + csv_name)
```
